```python
import numpy as np
import matplotlib.pyplot as plt

from mfpml.design_of_experiment.singlefideliy_samplers import FixNumberSampler, RandomSampler, LatinHyperCube, SobolSequence
```

```python
design_space = np.array([[1, 5], [-1, 1]])
```

```python
# FixNumberSampler
fix_sampler = FixNumberSampler(np.array([[1], [3]]))
fix_sampler.get_samples(num_samples=10)
```

```python
# RandomSampler
random_sampler = RandomSampler(design_space)
random_sampler.get_samples(num_samples=10)

# plot samples
fig, ax = plt.subplots()
ax.scatter(random_sampler.samples[:, 0], random_sampler.samples[:, 1])
ax.set_xlabel("x1")
ax.set_ylabel("x2")
ax.set_title("Random Sampler")
```

```python
# LatinHyperCube
lhs_sampler = LatinHyperCube(design_space)
lhs_sampler.get_samples(num_samples=10)

# plot samples
fig, ax = plt.subplots()
ax.scatter(lhs_sampler.samples[:, 0], lhs_sampler.samples[:, 1])
ax.set_xlabel("x1")
ax.set_ylabel("x2")
ax.set_title("Latin Hyper Cube Sampler")
plt.show()
```

```python
# SobolSequence
sobol_sampler = SobolSequence(design_space)
sobol_sampler.get_samples(num_samples=10)

# plot samples
fig, ax = plt.subplots()
ax.scatter(sobol_sampler.samples[:, 0], sobol_sampler.samples[:, 1])
ax.set_xlabel("x1")
ax.set_ylabel("x2")
ax.set_title("Sobol Sequence Sampler")
plt.show()
```

```python
from mfpml.design_of_experiment.multifidelity_samplers import MFLatinHyperCube

method = MFLatinHyperCube(design_space=np.array([[5., 10],
                                                [1, 2]]),
                          num_fidelity=4,
                          nested=True)
data = method.get_samples(num_samples=[5, 20])
```

```python
data
```

```python

# plot samples
fig, ax = plt.subplots()

ax.scatter(data[1][:, 0], data[1][:, 1], color="gray",
           marker="*", label="fidelity 2")
ax.scatter(data[0][:, 0], data[0][:, 1], marker='+', label="fidelity 1")
plt.legend()
ax.set_xlabel("x1")
ax.set_ylabel("x2")
ax.set_title("Multifidelity Latin Hyper Cube Sampler")
plt.show()
```
