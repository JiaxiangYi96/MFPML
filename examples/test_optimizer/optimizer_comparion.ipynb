```python
import os
import sys
import time

folder_path = "/home/yaga/Documents/GitHub/mfpml"
sys.path.insert(0, folder_path)


from mfpml.problems.sf_functions import Ackley
from mfpml.optimization.evolutionary_algorithms import DE, PSO

function = Ackley(num_dim=20)
design_space = function._input_domain

```

```python
# define the optimizer with ramdom strategy 
de_opt_rand = DE(num_gen=1000, num_pop=50,strategy="DE/rand/1/bin")
rand_results = de_opt_rand.run_optimizer(
    func=function.f,
    num_dim=function.num_dim,
    design_space=design_space,
    print_info=False,
)
# print(opt_results)
# pso_opt_rand.plot_optimization_history(figure_name="de_rand")
```

```python
# define the optimizer with ramdom strategy 
de_opt_best = DE(num_gen=1000, num_pop=50,strategy="DE/best/1/bin")
best_results = de_opt_best.run_optimizer(
    func=function.f,
    num_dim=function.num_dim,
    design_space=design_space,
    print_info=False,
)
# print(opt_results)
# pso_opt_best.plot_optimization_history(figure_name="de_best")
```

```python
# define the optimizer with ramdom strategy 
pso_opt = PSO(num_gen=1000, num_pop=50)
best_results = pso_opt.run_optimizer(
    func=function.f,
    num_dim=function.num_dim,
    design_space=design_space,
    print_info=False,
)
```

```python
import matplotlib.pyplot as plt 
fig, ax = plt.subplots()
ax.plot(de_opt_rand.gen_best, label='DE/random/1/bin')
ax.plot(de_opt_best.gen_best, label='DE/best/1/bin')
ax.plot(pso_opt.gen_best, label='pso')
plt.legend()
plt.show()
```
