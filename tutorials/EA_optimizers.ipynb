```python
import os
import sys
import time
from mfpml.problems.sf_functions import Ackley
from mfpml.optimization.evolutionary_algorithms import DE, PSO
function = Ackley(num_dim=6)
design_space = function._input_domain
```

```python
# define the optimizer with ramdom strategy
de_opt_rand = DE(num_gen=200, num_pop=100,
                 strategy="DE/rand/1/bin", crossover_rate=0.1)
rand_results = de_opt_rand.run_optimizer(
    func=function.f,
    num_dim=function.num_dim,
    design_space=design_space,
    print_info=False,
)
```

```python
# define the optimizer with ramdom strategy
de_opt_best = DE(num_gen=100, num_pop=20,
                 strategy="DE/best/1/bin", crossover_rate=0.5)
best_results = de_opt_best.run_optimizer(
    func=function.f,
    num_dim=function.num_dim,
    design_space=design_space,
    print_info=False,
)
```

```python
# define the optimizer with ramdom strategy
pso_opt = PSO(num_gen=100, num_pop=20)
best_results = pso_opt.run_optimizer(
    func=function.f,
    num_dim=function.num_dim,
    design_space=design_space,
    print_info=False,
)
```

```python
import matplotlib.pyplot as plt
fig, ax = plt.subplots()
ax.plot(de_opt_rand.gen_best, label='DE/random/1/bin')
ax.plot(de_opt_best.gen_best, label='DE/best/1/bin')
ax.plot(pso_opt.gen_best, label='pso')
plt.legend()
plt.show()
```

```python
import numpy as np
aa = de_opt_rand.gen_best
bb = de_opt_rand.responses
xx = np.linspace(0, aa.shape[0], aa.shape[0]+1, endpoint=True)
fig, ax = plt.subplots()
ax.plot(aa)
for xe, ye in zip(xx, bb):
    plt.scatter([xe]*ye.shape[0], ye.ravel())
# plt.ylim([0,50])
plt.show()
```
